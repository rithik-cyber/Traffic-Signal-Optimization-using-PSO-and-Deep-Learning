```python
import osmnx as ox
import matplotlib.pyplot as plt

# Fetch full Chandigarh road network
try:
    print("Fetching full Chandigarh road network...")
    full_graph = ox.graph_from_place("Chandigarh, India", network_type="drive")

    # Save the full network
    ox.save_graphml(full_graph, "chandigarh_full_network.graphml")

    # Plot the full network
    fig, ax = plt.subplots(figsize=(12, 12))
    ox.plot_graph(full_graph, ax=ax, node_size=5, edge_linewidth=0.7, edge_color="black", node_color="red")
    plt.title("Full Chandigarh Road Network")
    plt.show()

except Exception as e:
    print(f"⚠️ Error fetching full Chandigarh network: {e}")
    print("Switching to a smaller bounding box...")

# Fetch a smaller bounding box inside Chandigarh
try:
    print("Fetching a smaller region of Chandigarh...")
    
    # Define bounding box manually (North, South, East, West)
    north, south, east, west = 30.75, 30.70, 76.85, 76.80  # A smaller part of Chandigarh

    # ✅ Corrected function call - Pass BBOX as a SINGLE TUPLE
    small_graph = ox.graph_from_bbox((north, south, east, west), network_type="drive")

    # Save the smaller network
    ox.save_graphml(small_graph, "chandigarh_small_network.graphml")

    # Plot the smaller network
    fig, ax = plt.subplots(figsize=(10, 10))
    ox.plot_graph(small_graph, ax=ax, node_size=5, edge_linewidth=0.7, edge_color="black", node_color="red")
    plt.title("Smaller Area of Chandigarh Road Network")
    plt.show()

except Exception as e:
    print(f"⚠️ Error fetching smaller region: {e}")
```

```python
import pygame
import random
import osmnx as ox
import networkx as nx

# Initialize pygame
pygame.init()

# Constants
WIDTH, HEIGHT = 1000, 800
BACKGROUND_COLOR = (200, 200, 200)
ROAD_COLOR = (50, 50, 50)
CAR_COLOR = (0, 0, 255)
TRAFFIC_LIGHT_COLORS = {'red': (255, 0, 0), 'yellow': (255, 255, 0), 'green': (0, 255, 0)}
FPS = 30

# Load Sector 17 road network
graph = ox.graph_from_place("Sector 17, Chandigarh, India", network_type='drive')
nodes, edges = ox.graph_to_gdfs(graph, nodes=True, edges=True)

# Scale factor to fit map into screen
def scale_coordinates(x, y, min_x, min_y, scale_x, scale_y):
    return int((x - min_x) * scale_x), int((y - min_y) * scale_y)

min_x, min_y, max_x, max_y = nodes.total_bounds
scale_x, scale_y = WIDTH / (max_x - min_x), HEIGHT / (max_y - min_y)

# Initialize screen
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Sector 17 Traffic Simulation")

# Car class
class Car:
    def __init__(self, x, y, speed=2):
        self.x, self.y = x, y
        self.speed = speed
        self.color = CAR_COLOR

    def move(self):
        self.x += self.speed
        if self.x > WIDTH:
            self.x = 0

    def draw(self):
        pygame.draw.rect(screen, self.color, (self.x, self.y, 15, 30))

# Traffic Light class
class TrafficLight:
    def __init__(self, x, y):
        self.x, self.y = x, y
        self.state = 'red'
        self.timer = 0

    def update(self):
        self.timer += 1
        if self.timer > 100:
            self.state = 'green' if self.state == 'red' else 'red'
            self.timer = 0

    def draw(self):
        pygame.draw.circle(screen, TRAFFIC_LIGHT_COLORS[self.state], (self.x, self.y), 10)

# Initialize cars and traffic lights
cars = [Car(random.randint(0, WIDTH), random.randint(0, HEIGHT), random.randint(1, 3)) for _ in range(10)]
traffic_lights = [TrafficLight(random.randint(100, WIDTH-100), random.randint(100, HEIGHT-100)) for _ in range(5)]

# Main loop
running = True
clock = pygame.time.Clock()
while running:
    screen.fill(BACKGROUND_COLOR)
    
    # Draw roads
    for _, edge in edges.iterrows():
        x1, y1 = scale_coordinates(edge.geometry.coords[0][0], edge.geometry.coords[0][1], min_x, min_y, scale_x, scale_y)
        x2, y2 = scale_coordinates(edge.geometry.coords[-1][0], edge.geometry.coords[-1][1], min_x, min_y, scale_x, scale_y)
        pygame.draw.line(screen, ROAD_COLOR, (x1, y1), (x2, y2), 5)

    # Update and draw traffic lights
    for light in traffic_lights:
        light.update()
        light.draw()
    
    # Update and draw cars
    for car in cars:
        car.move()
        car.draw()
    
    # Event handling
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    
    pygame.display.flip()
    clock.tick(FPS)

pygame.quit()
```
