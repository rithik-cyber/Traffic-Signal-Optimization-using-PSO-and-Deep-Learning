```python
import pygame
from pygame.locals import *
from OpenGL.GL import *
from OpenGL.GLU import *
import random
import csv

# Initialize Pygame
pygame.init()

# Set display dimensions
WIDTH, HEIGHT = 1000, 800
screen = pygame.display.set_mode((WIDTH, HEIGHT), DOUBLEBUF | OPENGL)
pygame.display.set_caption("Sector 17 - 3D Traffic Simulation")

gluPerspective(45, (WIDTH / HEIGHT), 0.1, 50.0)
glTranslatef(0.0, -1.0, -10)

traffic_log = open("traffic_data.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(traffic_log, delimiter=",", quotechar='"', quoting=csv.QUOTE_MINIMAL)
csv_writer.writerow(["Time", "Traffic Light State", "Num Cars Waiting"])


def draw_road():
    """Draws a simple 3D road grid for Sector 17 with intersections"""
    glBegin(GL_QUADS)
    glColor3f(0.2, 0.2, 0.2)  # Gray road
    for x in range(-5, 6, 2):
        for z in range(-5, 6, 2):
            glVertex3f(x, 0, z)
            glVertex3f(x+1, 0, z)
            glVertex3f(x+1, 0, z+1)
            glVertex3f(x, 0, z+1)
    glEnd()

def draw_building(x, z):
    """Draws a basic building at a given position"""
    glBegin(GL_QUADS)
    glColor3f(0.5, 0.5, 0.5)
    for y in range(0, 5):
        glVertex3f(x-0.5, y, z-0.5)
        glVertex3f(x+0.5, y, z-0.5)
        glVertex3f(x+0.5, y, z+0.5)
        glVertex3f(x-0.5, y, z+0.5)
    glEnd()

def draw_traffic_light(x, z, state):
    """Draws a traffic light at intersections with timed state changes"""
    glColor3f(1, 0, 0) if state == 'red' else glColor3f(0, 1, 0)
    glBegin(GL_QUADS)
    glVertex3f(x-0.1, 1, z-0.1)
    glVertex3f(x+0.1, 1, z-0.1)
    glVertex3f(x+0.1, 2, z-0.1)
    glVertex3f(x-0.1, 2, z-0.1)
    glEnd()

def draw_car(x, z):
    """Draws a moving car at a given position"""
    glColor3f(1, 0, 0)  # Red car
    glBegin(GL_QUADS)
    glVertex3f(x-0.3, 0.1, z-0.5)
    glVertex3f(x+0.3, 0.1, z-0.5)
    glVertex3f(x+0.3, 0.1, z+0.5)
    glVertex3f(x-0.3, 0.1, z+0.5)
    glEnd()

def draw_pedestrian(x, z):
    """Draws a simple pedestrian"""
    glColor3f(1, 1, 0)
    glBegin(GL_QUADS)
    glVertex3f(x-0.1, 0, z-0.1)
    glVertex3f(x+0.1, 0, z-0.1)
    glVertex3f(x+0.1, 1, z-0.1)
    glVertex3f(x-0.1, 1, z-0.1)
    glEnd()

traffic_light_state = 'red'
traffic_light_timer = 0
car_positions = [[random.uniform(-4, 4), -4] for _ in range(3)]  # Three moving cars
pedestrians = [[random.uniform(-4, 4), random.uniform(-4, 4)] for _ in range(5)]
time_step = 0

def update_simulation():
    global traffic_light_state, traffic_light_timer, time_step, car_positions
    traffic_light_timer += 1
    time_step += 1
    
    # Count cars waiting at red light
    num_cars_waiting = sum(1 for car in car_positions if traffic_light_state == 'red' and car[1] < -2)
    csv_writer.writerow([time_step, traffic_light_state, num_cars_waiting])
    traffic_log.flush()  # Ensure immediate write to file
    
    print(f"Time: {time_step}, Traffic Light: {traffic_light_state}, Cars Waiting: {num_cars_waiting}")
    
    if traffic_light_timer > 50:
        traffic_light_state = 'green' if traffic_light_state == 'red' else 'red'
        traffic_light_timer = 0
    
    # Move existing cars and remove those that leave the screen
    new_car_positions = []
    for car in car_positions:
        if traffic_light_state == 'green':
            car[1] += 0.1  # Move car forward
        if car[1] < 5:  # Keep cars within bounds
            new_car_positions.append(car)
    
    # Add new cars at random intervals
    if random.random() < 0.3:  # 30% chance of a new car appearing
        new_car_positions.append([random.uniform(-4, 4), -4])
    
    car_positions = new_car_positions

def draw_scene():
    """Renders the entire scene"""
    glClear(GL_COLOR_BUFFER_BIT | GL_DEPTH_BUFFER_BIT)
    draw_road()
    for x in range(-4, 5, 2):
        for z in range(-4, 5, 2):
            draw_building(x, z)
    draw_traffic_light(0, -3, traffic_light_state)
    for car in car_positions:
        draw_car(car[0], car[1])
    for ped in pedestrians:
        draw_pedestrian(ped[0], ped[1])

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    update_simulation()
    draw_scene()
    pygame.display.flip()
    pygame.time.wait(100)

pygame.quit()
traffic_log.close()
```
